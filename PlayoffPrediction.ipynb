```python
import pandas as pd 
import time as time 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from nba_api.stats.endpoints import playercareerstats 
from nba_api.stats.static import players
%matplotlib inline
import plotly.express as px

result = pd.read_csv("/Users/hass/Desktop/1983-21LeagueStandingsAllFeaturesPerGame.csv")
# result

result.columns

def playoffPrediction(conference, year):
    # conference (string): "East"/"West"
    #year (int): 2018/2019/2020

    easternConf = result.loc[result["Conference"] == conference]

    train_2019 = easternConf.loc[easternConf["SeasonID"] < year]
    test_2019 = easternConf.loc[easternConf["SeasonID"] == year]

    X_train_2019 = train_2019[['WIN_PCT', 'FGM', 'FGA', 'FG_PCT',
           'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB',
           'REB', 'AST', 'PF', 'STL', 'TOV', 'BLK', 'PTS']]
    X_test_2019 = test_2019[['WIN_PCT', 'FGM', 'FGA', 'FG_PCT',
           'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB',
           'REB', 'AST', 'PF', 'STL', 'TOV', 'BLK', 'PTS']]

    y_train_2019 = train_2019[['playoffsBool']]
    y_test_2019 = test_2019[['playoffsBool']]

    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression

    print ("Logistic - East")
    lr_2019East = LogisticRegression().fit(X_train_2019, y_train_2019)
    lr_2019East_train_score = lr_2019East.score(X_train_2019, y_train_2019)
    lr_2019East_test_score = lr_2019East.score(X_test_2019, y_test_2019)
    print(f"Training Data Score: {lr_2019East_train_score}")
    print(f"Testing Data Score: {lr_2019East_test_score}")

    from sklearn.svm import SVC 
    print ("SVM - East")
    svm_2019East = SVC(kernel='linear', probability=True).fit(X_train_2019, y_train_2019)
    svm_2019East_train_score = svm_2019East.score(X_train_2019, y_train_2019)
    svm_2019East_test_score = svm_2019East.score(X_test_2019, y_test_2019)
    print(f"Training Data Score: {svm_2019East_train_score}")
    print(f"Testing Data Score: {svm_2019East_test_score}")

    from sklearn.neural_network import MLPRegressor
    from sklearn.linear_model import LinearRegression
    from sklearn.ensemble import RandomForestClassifier

    east_predictions_2019 = test_2019[["TeamCity", "TeamName", "SeasonID", "playoffsBool", "Conference"]]
    east_predictions_2019 = east_predictions_2019.loc[east_predictions_2019["SeasonID"] == year]
    east_predictions_2019

    lr_probability = lr_2019East.predict_proba(X_test_2019)[:,1].tolist()
    svm_probability = svm_2019East.predict_proba(X_test_2019)[:,1].tolist()

    lr_prediction = lr_2019East.predict(X_test_2019).tolist()
    svm_prediction = svm_2019East.predict(X_test_2019).tolist()

    east_predictions_2019["LogRegProb"] = lr_probability
    east_predictions_2019["SVMProb"] = svm_probability

    east_predictions_2019.sort_values("LogRegProb", ascending = False)
    east_predictions_2019.sort_values("SVMProb", ascending = False)

    high_pct = east_predictions_2019.nlargest(8, ["LogRegProb"]).iloc[7]["LogRegProb"]
    east_predictions_2019["LogRegBool"] = east_predictions_2019["LogRegProb"].map(lambda x: 1 if x >= high_pct else 0)

    high_pct = east_predictions_2019.nlargest(8, ["SVMProb"]).iloc[7]["SVMProb"]
    east_predictions_2019["SVMBool"] = east_predictions_2019["SVMProb"].map(lambda x: 1 if x >= high_pct else 0)

    return east_predictions_2019
```

```python
from nba_api.stats.static import teams
import pandas as pd 
from nba_api.stats.endpoints import playerestimatedmetrics
import plotly.express as px
from nba_api.stats.endpoints import commonteamroster

def teamAdvancedStats(teamName, year):
    # teams.find_teams_by_full_name('Lakers')
    value = teamName
    teamValue = teams.find_teams_by_full_name(value)
    playerId = []
    playerName = []
    offR = []
    defR = []
    netR = []
    usgPct = []
    if(len(teamValue) >= 1):
        print('team found')
        tempId = teamValue[0]['id']
        playerList = commonteamroster.CommonTeamRoster(season = year, team_id = tempId).get_data_frames()[0]
        for index, row in playerList.iterrows():
            playerId.append(row['PLAYER_ID'])
            playerName.append(row['PLAYER'])

        list_of_tuples = list(zip(playerId, playerName))
        df = pd.DataFrame(list_of_tuples,
                      columns = ['playerId', 'playerName'])
#         print(df)
        
        pem = playerestimatedmetrics.PlayerEstimatedMetrics(season = year)
        pem = pem.get_data_frames()[0]
        for index, row in df.iterrows():
            if(len(pem[pem['PLAYER_ID'] == int(row['playerId'])]) > 0):
                offR.append(pem[pem['PLAYER_ID'] == int(row['playerId'])].iloc[0]["E_OFF_RATING"])
                defR.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_DEF_RATING"])
                netR.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_NET_RATING"])
                usgPct.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_USG_PCT"])
            else:
                df.drop(index, inplace=True)
        df['offR'] = offR
        df['defR'] = defR
        df['netR'] = netR
        df['usgPct'] = usgPct
#         print(df)
        
        playerId = []
        playerName = []
        offR = []
        defR = []
        netR = []
        usgPct = []
        offR2 = []
        defR2 = []
        netR2 = []
        usgPct2 = []
        previousYear = str(int(year[0:4])-1) + year[4:5] + (str(int(year[5:7])-1))
        playerList = commonteamroster.CommonTeamRoster(season = previousYear, team_id = tempId).get_data_frames()[0]
        for index, row in playerList.iterrows():
            if(len(df[df['playerId'] == row['PLAYER_ID']] ) > 0):
                playerId.append(row['PLAYER_ID'])
                playerName.append(row['PLAYER'])
                offR.append(df[df['playerId'] == row['PLAYER_ID']].iloc[0]['offR'])
                defR.append(df[df['playerId'] == row['PLAYER_ID']].iloc[0]['defR'])
                netR.append(df[df['playerId'] == row['PLAYER_ID']].iloc[0]['netR'])
                usgPct.append(df[df['playerId'] == row['PLAYER_ID']].iloc[0]['usgPct'])
                
        list_of_tuples = list(zip(playerId, playerName, offR, defR, netR, usgPct))
        df2 = pd.DataFrame(list_of_tuples,
                      columns = ['playerId', 'playerName', 'offR', 'defR', 'netR', 'usgPct'])
        
        pem = playerestimatedmetrics.PlayerEstimatedMetrics(season = previousYear)
        pem = pem.get_data_frames()[0]
        for index, row in df2.iterrows():
            if(len(pem[pem['PLAYER_ID'] == int(row['playerId'])]) > 0):
                offR2.append(pem[pem['PLAYER_ID'] == int(row['playerId'])].iloc[0]["E_OFF_RATING"])
                defR2.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_DEF_RATING"])
                netR2.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_NET_RATING"])
                usgPct2.append(pem[pem['PLAYER_ID'] == row['playerId']].iloc[0]["E_USG_PCT"])
            else:
                df2.drop(index, inplace=True)
        df2['offR2'] = offR2
        df2['defR2'] = defR2
        df2['netR2'] = netR2
        df2['usgPct2'] = usgPct2
        
        return df, df2
        
#         fig1 = px.bar(df, 
#                   x=df["playerName"], 
#                   y=df["netR"],
#                   barmode='group',
#                   labels={
#                          "x": "Player",
#                          "y": "offRating" ,

#                      },
#                   title="Offensive Rating of Players"
#                  )
#         fig1.show()
#         fig2 = px.bar(df2, 
#                   x=df2["playerName"], 
#                   y=df2["netR"]- df2["netR2"],
#                   barmode='group',
#                   labels={
#                          "x": "Player",
#                          "y": "netRating" ,

#                      },
#                   title="Net Rating Difference of Players"
#                  )
#         fig2.show()
    else:
        print('nay')

```

```python
import dash_html_components as html

def discrete_background_color_bins(df, n_bins=5, columns='all'):
    import colorlover
    bounds = [i * (1.0 / n_bins) for i in range(n_bins + 1)]
    if columns == 'all':
        if 'id' in df:
            df_numeric_columns = df.select_dtypes('number').drop(['id'], axis=1)
        else:
            df_numeric_columns = df.select_dtypes('number')
    else:
        df_numeric_columns = df[columns]
    df_max = df_numeric_columns.max().max()
    df_min = df_numeric_columns.min().min()
    ranges = [
        ((df_max - df_min) * i) + df_min
        for i in bounds
    ]
    styles = []
    legend = []
    for i in range(1, len(bounds)):
        min_bound = ranges[i - 1]
        max_bound = ranges[i]
        backgroundColor = colorlover.scales[str(n_bins)]['seq']['Blues'][i - 1]
        color = 'white' if i > len(bounds) / 2. else 'inherit'

        for column in df_numeric_columns:
            styles.append({
                'if': {
                    'filter_query': (
                        '{{{column}}} >= {min_bound}' +
                        (' && {{{column}}} < {max_bound}' if (i < len(bounds) - 1) else '')
                    ).format(column=column, min_bound=min_bound, max_bound=max_bound),
                    'column_id': column
                },
                'backgroundColor': backgroundColor,
                'color': color
            })
        legend.append(
            html.Div(style={'display': 'inline-block', 'width': '60px'}, children=[
                html.Div(
                    style={
                        'backgroundColor': backgroundColor,
                        'borderLeft': '1px rgb(50, 50, 50) solid',
                        'height': '10px'
                    }
                ),
                html.Small(round(min_bound, 2), style={'paddingLeft': '2px'})
            ])
        )

    return (styles, html.Div(legend, style={'padding': '5px 0 5px 0'}))
```
